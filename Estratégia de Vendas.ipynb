```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Vendas.xlsx')
df.head()
```

```python
df.dtypes
```

```python
df['ID-Vendedor'] = df['ID-Vendedor'].astype('object')
```

# Total de vendas por fabricante

```python
df_fabricante = df.groupby('Fabricante').sum().sort_values('ValorVenda', ascending=False).reset_index()
```

```python
# df_fabricante.drop(columns='ID-Vendedor', inplace=True)
```

```python
df_fabricante
```

# Total de vendas por estado

```python
df.head()
```

```python
df.groupby('Estado').sum().sort_values('ValorVenda', ascending=False).reset_index()
```

# Total de Vendas por categoria

```python
df.head()
```

```python
df.groupby('Categoria').sum().sort_values('ValorVenda', ascending=False).reset_index()
```

# Segmento com maior influência no valor médio da venda

```python
df['ValorVenda'].mean()
```

```python
df.groupby('Segmento').mean().sort_values('ValorVenda', ascending=False).reset_index()
```

# Produto com maior influência no valor médio da venda

```python
df['ValorVenda'].mean()
```

```python
df.groupby('Produto').mean().sort_values('ValorVenda', ascending=False).reset_index()
```

# Total de vendas por vendedor

```python
df.head()
```

```python
df.groupby('Vendedor').sum().sort_values('ValorVenda', ascending=False).reset_index()
```

# Quantidade vendida por vendedor

```python
df['Vendedor'].value_counts()
```

# Total de vendas por estado e por categoria

```python
df.groupby(['Estado', 'Categoria']).sum().sort_values(['Estado', 'ValorVenda'], ascending=False).reset_index()
```

# Total de Vendas por ano

```python
df.head()
```

```python
df.groupby(pd.Grouper(key='Data Venda', freq='Y')).sum().sort_values('ValorVenda', ascending=False).reset_index()
```
